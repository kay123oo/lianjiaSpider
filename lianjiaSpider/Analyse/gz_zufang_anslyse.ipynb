```python
import pymongo
import pandas as pd
import numpy as np
```

```python
from pymongo import MongoClient

conn = MongoClient(host='127.0.0.1', port=27017)  # 实例化MongoClient
db = conn.get_database('lianjia')  # 连接到Lianjia数据库

zufang = db.get_collection('zufang_gz') # 连接到集合zufang
mon_data = zufang.find()  # 查询这个集合下的所有记录
```

```python
lianjia_df=pd.DataFrame(list(mon_data))
```

```python
lianjia_df.head(5)
```

```python
df=lianjia_df.copy()
df=df.drop(columns='_id')
df=df.drop(columns='title')
```

```python
df['perPrice']=df['price']/df['area']
df['hall']=df['houseType'].str[0]
df['bed']=df['houseType'].str[2]
df['bath']=df['houseType'].str[4]
col = ['city','district','microdistrict','community','price',
       'perPrice','area','orientation','orientation','tags','hall','bed','bath']
df
```

```python

df_house_count=df.loc[(df['dataDistributionType']==0),:].groupby('district')['price'].count().sort_values(ascending=False).to_frame().reset_index()
df_apartment_count=df.loc[(df['dataDistributionType']==1),:].groupby('district')['price'].count().sort_values(ascending=False).to_frame().reset_index()
df_house_mean=df.loc[(df['dataDistributionType']==1),:].groupby('district')['perPrice'].mean().round(1) .sort_values(ascending=False).to_frame().reset_index()
df_apartment_mean=df.loc[(df['dataDistributionType']==0),:].groupby('district')['perPrice'].mean().round(1) .sort_values(ascending=False).to_frame().reset_index()
print(df_house_mean)
```
